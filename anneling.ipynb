```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import anneling_process as ap
import plot_functions as fpl

r42 = np.random.RandomState(42)
```

```python
N = 100 # Колличество городов
T_init = 10  # Зададим начальную температуру
T_end = 0.0001 # Зададим конечную температуру

init_state = np.random.permutation(N) # Зададим рандомный маршрут между городами 
                                      # (последовательность номеров городов)
    
cities_coord = r42.rand(2,N)*10 # Зададим рандомные координаты городов

E_hist, T_hist, final_state = ap.SimulatedAnnealing(cities_coord, T_init, T_end) # Запустим алгоритм иммитации отжига
```

```python
fpl.plot_hist(E_hist, T_hist) # Построим зависимость изменения длины маршрута в процесси отжига
                              # А так же зависимость температуры отжига от времени
```

```python
fpl.plot_graph(cities_coord,init_state) # Визуализируем изначальный маршрут
```

```python
fpl.plot_graph(cities_coord,final_state, Title = 'Путь коммивояжера после оптимизации') # Визуализируем маршрут после оптимизации
```
